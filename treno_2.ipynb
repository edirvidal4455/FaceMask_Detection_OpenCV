```python
import cv2
import os
import numpy as np
```

```python
dataPath = "C:/Users/edirv/Desktop/Mascarillas dataset/Dataset_faces_2" #dirección o path donde se encuentran las imagenes para entrenar
dir_list = os.listdir(dataPath) #dirección de trabajo
print("Lista archivos:", dir_list) #muestra los archivos del directorio
```

```python
labels = []
facesData = []
label = 0
```

```python
#revisa cada imagen que está en el directorio
for name_dir in dir_list:
     dir_path = dataPath + "/" + name_dir #define la carpeta dentro del directorio a analizar (1 por vez)
     #revisa cada imagen que está en la carpeta
     for file_name in os.listdir(dir_path):
          image_path = dir_path + "/" + file_name
          #print(image_path) #revisa que todo vaya bien
          image = cv2.imread(image_path, 0)
          #cv2.imshow("Image", image) #no activar por que crashea xd en serio necesito más RAM para algo tan simple? bruh
          #cv2.waitKey(10) #no activar por que crashea xd en serio necesito más RAM para algo tan simple? bruh
          facesData.append(image) #inserta cada imagen en facesData
          labels.append(label) #inserta cada etiqueta en labels
     label += 1 #hace que todos los rostros que usan mascarilla son 0, los rostros que usan mal la mascarilla son 1, y los rostros que no usan mascarilla son 2

print("Etiqueta 0: ", np.count_nonzero(np.array(labels) == 0)) #cantidad de fotos con etiqueta 0 en labels
print("Etiqueta 1: ", np.count_nonzero(np.array(labels) == 1)) #cantidad de fotos con etiqueta 1 en labels
print("Etiqueta 2: ", np.count_nonzero(np.array(labels) == 2)) #cantidad de fotos con etiqueta 2 en labels
```

```python
# LBPH FaceRecognizer
face_mask = cv2.face.LBPHFaceRecognizer_create() #reconocimiento de rostros (no pienso programar 50 lineas para un programa que ya existe)
# Entrenamiento
print("Entrenando...") #notificación de que está entrenando UwU
face_mask.train(facesData, np.array(labels)) #en serio estas bibliotecas son lo máximo, entrenamiento con deep learning de OpenCV
face_mask.write("face_mask_model_2.xml") #crea un archivo .xml para almacenar el modelo entrenado
print("Modelo almacenado") #notificación de EXITO ABSOLUTO LETS GOOOO
#tarde 2 horas en este segmento, instalando y reinstalando opencv-contrib, hasta encontrar la version que tenía el método 'face'
```
